```python
import os
from datetime import datetime
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

class Inspection:
    def __init__(self, ot_number):
        self.ot_number = ot_number
        self.data = {}

    def start_inspection(self):
        self.data['Instalacao'] = input("Tipo de Instalação (monofasica/trifasica): ")
        self.data['Contrato_Ativo'] = input("A instalação possui contrato ativo? (Sim/Nao): ")
        self.data['Medicoes_Feitas'] = input("Foram feitas medições? (Sim/Nao): ")
        self.data['Numero_Serie_Contador'] = input("Quatro últimos dígitos do número de série do contador: ")
        self.data['Data'] = input("Data da inspeção (dd/mm/yyyy): ")
        self.data['Correcao'] = input("A inspeção foi sem correção (SC) ou com correção (CC): ")
        self.data['A_Vazio'] = input("Consumo no horário Vazio: ")
        self.data['A_Ponta'] = input("Consumo no horário de ponta: ")
        self.data['A_Cheia'] = input("Consumo no horário cheio: ")

        if self.data['Medicoes_Feitas'].lower() == 'sim':
            if self.data['Instalacao'].lower() == 'monofasica':
                self.data['Tensao'] = input("Valor da tensão (U): ")
                self.data['Corrente'] = input("Valor da corrente (I): ")
            elif self.data['Instalacao'].lower() == 'trifasica':
                self.data['U1'] = input("Valor da tensão da fase 1 (U1): ")
                self.data['U2'] = input("Valor da tensão da fase 2 (U2): ")
                self.data['U3'] = input("Valor da tensão da fase 3 (U3): ")
                self.data['I1'] = input("Valor da corrente da fase 1 (I1): ")
                self.data['I2'] = input("Valor da corrente da fase 2 (I2): ")
                self.data['I3'] = input("Valor da corrente da fase 3 (I3): ")

        self.data['Selo_retirado_Bornes'] = input("Número do selo retirado nos bornes do contador: ")
        self.data['Selo_retirado_Tampa'] = input("Número do selo retirado na tampa da caixa do contador: ")
        self.data['Selo_retirado_Portinhola'] = input("Número do selo retirado na portinhola: ")
        self.data['Selo_Colocado_Bornes'] = input("Número do selo colocado nos bornes do contador: ")
        self.data['Selo_Colocado_Tampa'] = input("Número do selo colocado na tampa da caixa do contador: ")
        self.data['Selo_Colocado_Portinhola'] = input("Número do selo colocado na portinhola: ")

        # Pergunta o número do selo colocado apenas se um selo foi retirado
        #if self.data['Selo_Bornes'] or self.data['Selo_Tampa'] or self.data['Selo_Portinhola']:
            #self.data['Selo_Colocado'] = input("Número do selo colocado: ")


    def generate_txt(self):
        formatted_date = datetime.strptime(self.data['Data'], "%d/%m/%Y").strftime("%Y%m%d")
        file_path = f"{self.ot_number}_{formatted_date}.txt"

        with open(file_path, 'w', encoding='utf-8') as txt_file:
            for key, value in self.data.items():
                txt_file.write(f"{key}: {value}\n")

        return file_path

def send_email(files, subject):
    smtp_server = "smtp.gmail.com"
    smtp_port = 587
    smtp_username = "patrickahotondji@gmail.com"  # Substitua pelo seu e-mail
    smtp_password = "nqyc vbfd zbkv jpbl"  # Substitua pela sua senha

    sender_email = "patrickahotondji@gmail.com"  # Substitua pelo seu e-mail
    receiver_email = "pma@iep.pt"  # Substitua pelo e-mail do destinatário

    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = subject

    body = "Segue em anexo os relatórios de inspeções."
    msg.attach(MIMEText(body, 'plain'))

    for file in files:
        with open(file, "rb") as attachment:
            part = MIMEText(attachment.read(), 'plain', 'utf-8')
            part.add_header('Content-Disposition', f'attachment; filename= {file}')
            msg.attach(part)

    with smtplib.SMTP(smtp_server, smtp_port) as server:
        server.starttls()
        server.login(smtp_username, smtp_password)
        server.sendmail(sender_email, receiver_email, msg.as_string())

def main():
    inspections = []
    txt_files = []

    while True:
        new_ot = input("Digite 'Nova OT' para começar uma nova ordem de trabalho (ou 'Sair' para sair): ")

        if new_ot.lower() == 'sair':
            break

        ot_number = input("Número da Ordem de Trabalho (N° OT): ")
        inspection = Inspection(ot_number)
        inspection.start_inspection()
        file_path = inspection.generate_txt()
        inspections.append(inspection)
        txt_files.append(file_path)

    send_email(txt_files, "Relatório de Inspeções")

if __name__ == "__main__":
    main()
```
